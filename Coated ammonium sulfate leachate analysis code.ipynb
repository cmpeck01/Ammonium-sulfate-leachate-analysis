```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('AS leachate data.csv')

# Convert 'Date' to datetime format with specified format
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%y")

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Unique trials and compounds
trials = df['Sample'].unique()
compounds = {
    'TKN mg': 'blue',
    'NH4 mg': 'green',
    'NO3 mg': 'red'
}

# Create separate plots for each compound and trial
for compound, color in compounds.items():
    for trial in trials:
        trial_data = df[df['Sample'] == trial][compound]
        
        # Check if there are any valid data points for the trial and compound
        if trial_data.notna().any():
            plt.figure(figsize=(12, 6))
            plt.plot(trial_data.index, trial_data, marker='o', linestyle='-', color=color, label=trial)
            
            # Formatting the plot
            plt.title(f'{compound} {trial}', fontsize=16)
            plt.xlabel('Date', fontsize=14)
            plt.ylabel('Mg leached', fontsize=14)
            plt.xticks(rotation=45)
            plt.legend(title='Trial', fontsize=12)
            plt.grid()
            plt.tight_layout()
            
            # Show the plot
            plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('AS leachate data.csv')

# Convert 'Date' to datetime format with specified format
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%y")

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Define the experimental groups
experimental_groups = ['AS-1', 'AS-2', 'AS-3']  # AS 1 and AS 2
compounds = ['TKN mg', 'NH4 mg', 'NO3 mg']
colors = ['blue', 'green', 'red']  # Define colors for each compound

# Prepare a DataFrame to store averages
averaged_data = df[df['Sample'].isin(experimental_groups)].groupby(['Date'])[compounds].mean()

# Create separate plots for each compound based on the average mg values
for compound, color in zip(compounds, colors):
    plt.figure(figsize=(12, 6))
    
    # Plot the averages
    plt.plot(averaged_data.index, averaged_data[compound], marker='o', linestyle='-', color=color, label=f'Average {compound}')
    
    # Formatting the plot
    plt.title(f'Average {compound} for AS 1,2,3', fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Average mg', fontsize=14)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the experimental groups
experimental_groups_cassicn9 = ['CAS-SiCN9-1', 'CAS-SiCN9-2', 'CAS-SiCN9-3']  
experimental_groups_as = ['AS-1', 'AS-2', 'AS-3']
experimental_groups_cassi = ['CAS-Si-1', 'CAS-Si-2', 'CAS-Si-3']
compounds = ['NH4 mg', 'NO3 mg', 'TKN mg']  # List of compounds to analyze

# Define colors for each treatment
colors = ['purple', 'red', 'green']

# Function to calculate total sums for each trial across all dates
def calculate_trial_totals(df, groups):
    trial_totals = {group: {} for group in groups}
    
    for group in groups:
        group_data = df[df['Sample'] == group]
        for compound in compounds:
            total_sum = group_data[compound].sum()
            trial_totals[group][compound] = total_sum
    
    return trial_totals

# Calculate total sums for each group
total_sum_cassicn9 = calculate_trial_totals(df2, experimental_groups_cassicn9)
total_sum_cassi = calculate_trial_totals(df3, experimental_groups_cassi)
total_sum_as = calculate_trial_totals(df, experimental_groups_as)

# Function to calculate averages and min/max for each analyte
def calculate_averages_and_min_max():
    averages = {}
    mins = {}
    maxs = {}
    
    for compound in compounds:
        sums_as = [total_sum_as[group][compound] for group in experimental_groups_as]
        sums_cassi = [total_sum_cassi[group][compound] for group in experimental_groups_cassi]
        sums_cassicn9 = [total_sum_cassicn9[group][compound] for group in experimental_groups_cassicn9]
        
        avg_as = np.mean(sums_as)
        avg_cassi = np.mean(sums_cassi)
        avg_cassicn9 = np.mean(sums_cassicn9)

        averages[compound] = [avg_as, avg_cassi, avg_cassicn9]
        mins[compound] = [np.min(sums_as), np.min(sums_cassi), np.min(sums_cassicn9)]
        maxs[compound] = [np.max(sums_as), np.max(sums_cassi), np.max(sums_cassicn9)]
    
    return averages, mins, maxs

# Calculate averages, mins, and maxs
averages, mins, maxs = calculate_averages_and_min_max()

# Create a single figure with subplots for each analyte
fig, axs = plt.subplots(1, len(compounds), figsize=(18, 6))  # 1 row, 3 columns

# Create plots for each compound
for i, compound in enumerate(compounds):
    # Create a bar graph for averages
    axs[i].bar(np.arange(len(colors)), averages[compound], width=0.35, color=colors)

    # Add error bars using mins and maxs
    axs[i].errorbar(np.arange(len(colors)), averages[compound], 
                    yerr=[np.array(averages[compound]) - np.array(mins[compound]), 
                          np.array(maxs[compound]) - np.array(averages[compound])], 
                    fmt='none', color='black', capsize=5)

    # Formatting the subplot
    axs[i].set_title(f'Average Cumulative Leached {compound}', fontsize=20)
    axs[i].set_xticks(np.arange(len(colors)))
    axs[i].set_xticklabels(['AS', 'CAS-Si', 'CAS-SiCN9'], fontsize=18)
    axs[i].set_ylabel('Total mg Released', fontsize=18)
    axs[i].tick_params(axis='both', which='major', labelsize=18)
    axs[i].grid()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
